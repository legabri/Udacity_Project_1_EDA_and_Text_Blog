### At this point I created an auxiliary variable called 'YEAR' and a summary df that will be used later to calculate the birth variations between the years

```python
df_provincial['YEAR']=df_provincial.PERIODE.str.replace(' ','')
df_provincial['YEAR']=df_provincial.PERIODE.str.replace('/','')
df_provincial['YEAR']=df_provincial.PERIODE.str.replace('-','')
df_provincial['YEAR']=df_provincial.PERIODE.str[5:10]

df_provincial
```

### The data format was a string, I converted them to int because I will do the math in the next steps

```python
summary_table['YEAR'] = summary_table['YEAR'].astype(int)
summary_table['NBR_ACCOUCH_ET_CESARIEN'] = summary_table['NBR_ACCOUCH_ET_CESARIEN'].astype(int)
```

```python
#Now we go calculate some variations between the periods
```

```python
def variation_between_periords(initial_period,final_period):
    
    '''this fonction calculate the variation of number of births in percentage between 2 periods and show the result
    input(2 values): initial period and final period
    output(1 value): percentual of variation between two given periods with 2 decimal numbers
    '''
          
    initial_value = summary_table.loc[summary_table.YEAR == initial_period, 'NBR_ACCOUCH_ET_CESARIEN'].astype(int)
    final_value = summary_table.loc[summary_table.YEAR == final_period, 'NBR_ACCOUCH_ET_CESARIEN'].astype(int)
    #print(initial_value,final_value)
    var = round(((final_value/initial_value)-1)*100,2).astype(float)
    #return var,initial_value
    #a=initial_value+final_value
    #print(a)
    if var.item() > 0:
        print('Variation between the periods:',
              '\n ',initial_period,'-',final_period,': Increase',var.item(),'%')
    elif var.item()== 0:
        print('Variation between the periods:',
              '\n ',initial_period,'-',final_period,': Stable',var.item(),'%')
    else:
        print('Variation between the periods:',
              '\n ', initial_period,'-',final_period,'=> Decrease',abs(var.item()),'%')
```
