```python
df2=df1[['NEW_TYPE','REGION','NBR_ACCOUCH_ET_CESARIEN','NBR_CESARIENNES','NBR_NAISS_VIVANT','NBR_MORTINAISSANC','PERIODE']]
```

```python
#note 1 = les abandons ne sont pas comptés
df2.head()
```

```python
# births by regions 
df_only_regions=df2.loc[df2['REGION']!='TOTAL PROVINCIAL']

#aggregate provincial births

df_provincial=df2.loc[df2['REGION']=='TOTAL PROVINCIAL']
```

```python
# 10 highest cesarean rates by period and region in %


'''creation of the new variable '% CESARIENNES' made with the division of 2 existing variables
'''
df_only_regions['%CESARIENNES']=round(df_only_regions['NBR_CESARIENNES']/df_only_regions['NBR_ACCOUCH_ET_CESARIEN']*100,2)

'''creation of the new variable '% MORTINAISSANC' made with the division of 2 already existing variables.
Will be used in the next steps'''

df_only_regions['%MORTINAISSANC']=round(df_only_regions['NBR_MORTINAISSANC']/df_only_regions['NBR_ACCOUCH_ET_CESARIEN']*100,2)

'''show in descending order'''
df_only_regions[['PERIODE','REGION','NBR_ACCOUCH_ET_CESARIEN','%CESARIENNES']].sort_values(by='%CESARIENNES',ascending=False)[0:10]



```

```python
# 10 lowest cesarean rates by region and period in %


df_only_regions[['PERIODE','REGION','NBR_ACCOUCH_ET_CESARIEN','%CESARIENNES']].sort_values(by='%CESARIENNES',ascending=True)[0:10]
```

```python
# 10 highest mortality rates by period and region in %

df_only_regions[['PERIODE','REGION','NBR_ACCOUCH_ET_CESARIEN','%MORTINAISSANC']].sort_values(by='%MORTINAISSANC',ascending=False)[0:10]
```

```python
# 10 lowest mortality rates by period and region in %

df_only_regions[['PERIODE','REGION','NBR_ACCOUCH_ET_CESARIEN','%MORTINAISSANC']].sort_values(by='%MORTINAISSANC',ascending=True)[0:10]
```

```python
#Over all in province by period

df_provincial['%CESARIENNES']=round(df_provincial['NBR_CESARIENNES']/df_provincial['NBR_ACCOUCH_ET_CESARIEN']*100,2)
df_provincial['%MORTINAISSANC']=round(df_provincial['NBR_MORTINAISSANC']/df_provincial['NBR_ACCOUCH_ET_CESARIEN']*100,2)


df_provincial[['PERIODE','REGION','NBR_ACCOUCH_ET_CESARIEN','%MORTINAISSANC','%CESARIENNES']].sort_values(by='PERIODE',ascending=False)
```
